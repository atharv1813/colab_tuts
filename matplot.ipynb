```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 10, 2)
y = [i**2 for i in x]

plt.plot(x, y)
plt.title("SIMPLE LINE PLOT")
plt.xlabel("X-AXIS")
plt.ylabel("Y-Axis")

# Annotate each point with its coordinates
for i in range(len(x)):
    plt.annotate(f"({x[i]}, {y[i]})", (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
```

```python
from matplotlib import pyplot as plt

years=["2017", "2018", "2019", "2020"]
placements=[150, 275,324,295]


cse=[75,80,47,82]
it=[46,58, 35,52]
ece=[67,59,72,62]

ece_start=[cse[i]+it[i] for i in range(len(cse))]

plt.title("STACKED BAR CHART")
plt.bar(years,cse,width=0.4,color="black")
plt.bar(years, it, bottom=cse,width=0.4, color="green")
plt.bar(years,ece,bottom=ece_start,width=0.4,color="red")

plt.ylim(0, 225)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

years = ['2018', '2019', '2020', '2021']
cs = [50, 90, 85, 95]
it = [75, 85, 80, 92]
ece = [60, 55, 70, 85]
w = 0.2

plt.figure(figsize=(10,7))
cs_bar=np.arange(len(years))
it_bar=[i+w for i in cs_bar]
ece_bar=[i+w for i in it_bar]

plt.bar(cs_bar,cs,width=w, label="CSE")
plt.bar(it_bar,it,width=w, label="IT")
plt.bar(ece_bar,ece,width=w, label="ECE")

plt.xticks(cs_bar+w, years)
plt.ylim(0,100)
plt.title("PLACEMENT COMPARISION")
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt

student_performance=["Excellent", "Good", "Average", "Poor"]
student_values=[15,25,12,8]
plt.figure(figsize=(9,10))

plt.pie(student_values,autopct="%2.1f%%",labels=student_performance, explode=[0,0.1,0.2,0.44], shadow=True, colors=["blue", "black", "red", "purple"])
plt.legend(title="PERFORMANCE")
plt.show()
```

```python
from matplotlib import pyplot as plt

marks=[90,50,40,60,55,44,30,10,34,84]
grade_intervals=[0,35, 70,100]
plt.title("STUDENT GRADE")
plt.hist(marks,grade_intervals, histtype="bar", rwidth=0.7) # rwidth not just width
plt.xticks([0,35,72,102])
plt.xlabel("Percentage")
plt.ylabel("No. of Students")
plt.show()
```
