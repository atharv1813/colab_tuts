**TensorFlow** is an open-source machine learning framework developed by Google. It provides a comprehensive ecosystem for building and deploying machine learning and deep learning models. TensorFlow supports a wide range of functionalities, including low-level operations for numerical computation and high-level abstractions to simplify model building. ### Why Keras is Needed
**Keras** is a high-level API built on top of TensorFlow (since TensorFlow 2.0, Keras is integrated into TensorFlow itself as `tf.keras`). Keras simplifies the process of building and training deep learning models. While TensorFlow provides powerful, low-level capabilities, it can be complex and cumbersome to work with directly. Keras abstracts much of this complexity, offering user-friendly and intuitive interfaces for creating neural networks.

### Key Differences:
1. **Ease of Use**: Keras provides a simpler, more readable syntax, making it easier to build and train models compared to low-level TensorFlow code.
2. **Rapid Prototyping**: Keras is designed for quick experimentation, allowing users to prototype and iterate on models faster.
3. **Customizability**: If you need to fine-tune or optimize your model at a low level, you can still use TensorFlow’s full features alongside Keras.

### Use Case:
- **Keras** is ideal for beginners or for projects that require quick implementation and testing of deep learning models.
- **TensorFlow** is useful when you need more control over model architecture or when building highly optimized models for production.

In summary, Keras is built on top of TensorFlow to provide an easier interface, while TensorFlow itself offers the power and flexibility for more advanced customization when needed.

### Model Building in Keras
Keras provides two main ways to build models: **Sequential API** and **Functional API**.

---

### 1. **Sequential API**
The **Sequential API** is the simplest way to build a model in Keras. It is designed for models that are a linear stack of layers, where each layer has exactly one input tensor and one output tensor.

- **Structure**: Layers are added in sequence, from input to output, without branching or merging. Each layer feeds into the next layer.
- **Use Case**: It is ideal for simple neural networks like plain feed-forward networks, convolutional neural networks (CNNs) for image classification, or simple recurrent neural networks (RNNs).

**Example of Models that Fit Sequential API**:
- A model with an input layer, followed by a hidden layer, and then an output layer. Each layer is connected one after another.
- Simpler architectures where you do not need any complex connectivity between layers.

**Limitations of Sequential API**:
- It cannot handle models with multiple inputs or outputs.
- You cannot create models where layers connect in ways other than sequentially (like sharing layers or creating non-linear topologies).

---

### 2. **Functional API**
The **Functional API** is more flexible and powerful. It allows you to build models that are more complex, such as those with:
- Multiple inputs and outputs.
- Layers that connect in a non-linear way (e.g., shared layers, skip connections, or multi-branch architectures).

**Key Concepts**:
- **Directed Acyclic Graph (DAG)**: The model architecture can be represented as a graph of layers, where layers can have multiple inputs and outputs.
- **Flexibility**: You can design a wider variety of models, like models for tasks that require advanced architectures (e.g., residual networks, Inception models, or attention mechanisms).

**Use Cases for Functional API**:
- **Multi-Input and Multi-Output Models**: For example, a model that takes an image and text as input and generates a prediction.
- **Shared Layers**: Useful in models where you need to share layers across different paths, such as in Siamese networks for tasks like similarity learning.
- **Complex Topologies**: Architectures where you have skip connections, like in ResNet, or where multiple layers merge.

---

### Comparison: Sequential vs. Functional
- **Simplicity**: The Sequential API is simpler and faster to set up, perfect for straightforward models. The Functional API requires a bit more effort but provides a lot more flexibility.
- **Use Case**: Use the Sequential API when your model is a simple stack of layers. Use the Functional API when you need to build non-linear, multi-branch models.

### Advantages of Keras
1. **High-Level API**: Simplifies the process of building and training neural networks.
2. **Community and Support**: It has a large, active community and excellent documentation, making troubleshooting easier.
3. **Integration with TensorFlow**: Starting from TensorFlow 2.0, Keras is the official high-level API of TensorFlow, combining Keras's ease of use with TensorFlow's robustness.

### Conclusion
Keras offers a simple interface for building models while also supporting complex model architectures. The choice between the Sequential and Functional APIs depends on the complexity of your neural network design and your specific needs.

**1.Keras sequential model code**

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(10,)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()

print("\nModel architecture: ")
print(model)

from keras.utils import plot_model
plot_model(model, show_shapes = True)
```

**1.Keras functional model code**

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Define the input layer
input_layer = keras.Input(shape=(10,))

# Define some shared hidden layers
x = layers.Dense(64, activation='relu')(input_layer)
x = layers.Dense(32, activation='relu')(x)

# Define two separate output layers
output_1 = layers.Dense(1, activation='sigmoid', name='output_1')(x)
output_2 = layers.Dense(1, activation='sigmoid', name='output_2')(x)

# Create the model with inputs and outputs
model = keras.Model(inputs=input_layer, outputs=[output_1, output_2])

# Compile the model
model.compile(
    optimizer='adam',
    loss={'output_1': 'binary_crossentropy', 'output_2': 'binary_crossentropy'},
    metrics=['accuracy']
)

# Print model summary
model.summary()


from keras.utils import plot_model
plot_model(model, show_shapes = True)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# First input and its own architecture
input_1 = keras.Input(shape=(10,), name='input_1')
x1 = layers.Dense(32, activation='relu')(input_1)
x1 = layers.Dense(16, activation='relu')(x1)

# Second input and its own architecture
input_2 = keras.Input(shape=(8,), name='input_2')
x2 = layers.Dense(32, activation='relu')(input_2)
x2 = layers.Dense(16, activation='relu')(x2)

# Concatenate the two architectures
merged = layers.Concatenate()([x1, x2])

# Shared architecture after concatenation
shared = layers.Dense(64, activation='relu')(merged)
shared = layers.Dense(32, activation='relu')(shared)

# Final output layer
output = layers.Dense(1, activation='sigmoid', name='output')(shared)

# Define the model
model = keras.Model(inputs=[input_1, input_2], outputs=output)

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Print model summary
model.summary()

from keras.utils import plot_model
plot_model(model, show_shapes = True)
```

```python
# Same above code using functional api

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
from keras.utils import plot_model

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Reshape the data and normalize
x_train = x_train.reshape(-1, 28 * 28).astype("float32") / 255.0
x_test = x_test.reshape(-1, 28 * 28).astype("float32") / 255.0

# Define the input layer
inputs = keras.Input(shape=(28 * 28,))

# Define the hidden layers
x = layers.Dense(512, activation='relu')(inputs)
x = layers.Dense(256, activation='relu')(x)

# Define the output layer
outputs = layers.Dense(10)(x)

# Create the model
model = keras.Model(inputs=inputs, outputs=outputs)

# Print the model summary
print(model.summary())

# Plot the model architecture
plot_model(model, show_shapes=True)

# Compile the model
model.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    metrics=["accuracy"]
)

# Train the model
model.fit(x_train, y_train, batch_size=32, epochs=5, verbose=2)

# Evaluate the model on the test data
model.evaluate(x_test, y_test, batch_size=32, verbose=2)
```
